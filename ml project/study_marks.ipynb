```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

data={
    "studyhours":[1,2,3,4,5,6],
    "marks":[30,40,50,60,70,80]
}

df=pd.DataFrame(data)

x=df[["studyhours"]]
y=df["marks"]

model=LinearRegression()
model.fit(x,y)

prediction=model.predict([[7]])
print("predicted marks:",prediction)
```
